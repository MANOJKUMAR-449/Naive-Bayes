```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation
import seaborn as sns
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
from sklearn.preprocessing import LabelEncoder
```

```python
data2 = pd.read_csv('SalaryData_Test.csv')
data1 = pd.read_csv('SalaryData_Train.csv')
```

```python
data1.head()
```

```python
data1 = data1.drop(columns=['maritalstatus','relationship','race','sex'],axis=1)
```

```python
data1.head()
```

```python
data2 = data2.drop(columns=['maritalstatus','relationship','race','sex'],axis=1)
```

```python
labelencoder = LabelEncoder()
```

```python
data1.iloc[:,-1] = labelencoder.fit_transform(data1.iloc[:,-1])
```

```python
data1.head()
```

```python
data2.iloc[:,-1] = labelencoder.fit_transform(data2.iloc[:,-1])
```

```python
x_test = pd.get_dummies(data2.iloc[:,:9])
y_test = data2[['Salary']]
```

```python
x_test.head()
```

```python
y_test.head()
```

```python
x_train = pd.get_dummies(data1.iloc[:,:9])
y_train = data1[['Salary']]
```

```python
x_train.head()
```

```python
y_train.head()
```

```python
model=GaussianNB()
model=model.fit(x_train,y_train)
```

```python
y_test_pred=model.predict(x_test)
```

```python
metrics.accuracy_score(y_test,y_test_pred)
```

```python
y_train_pred=model.predict(x_train)
```

```python
metrics.accuracy_score(y_train,y_train_pred)
```

```python
model2 = MultinomialNB()
```

```python
model2.fit(x_train,y_train)
```

```python
y_train_pred=model2.predict(x_train)
```

```python
metrics.accuracy_score(y_train,y_train_pred)
```

```python
y_test_pred=model2.predict(x_test)
```

```python
metrics.accuracy_score(y_test,y_test_pred)
```

```python
model3 = BernoulliNB()
```

```python
model3.fit(x_train,y_train)
```

```python
y_train_pred=model3.predict(x_train)
```

```python
metrics.accuracy_score(y_train,y_train_pred)
```

```python
y_test_pred=model2.predict(x_test)
```

```python
metrics.accuracy_score(y_test,y_test_pred)
```
